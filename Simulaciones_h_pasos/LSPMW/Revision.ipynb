```python
from pipeline import PipelineUnified_LSPMW

# EJEMPLO DE USO
if __name__ == "__main__":
    # Crear pipeline unificado solo para LSPMW
    pipeline = PipelineUnified_LSPMW(seed=42, verbose=True)
    
    # Ejecutar todos los procesos (ARMA + ARIMA + SETAR)
    # Total: (7 ARMA + 7 ARIMA + 7 SETAR) √ó 5 distribuciones √ó 4 varianzas = 420 escenarios
    resultados = pipeline.run_all(
        excel_filename="resultados_LSPMW_todos_procesos.xlsx",
        n_jobs=4,
        batch_size=10
    )
    
    # An√°lisis por tipo de proceso
    print("\nüìä AN√ÅLISIS POR TIPO DE PROCESO:")
    print(resultados.groupby(['Tipo_Proceso', 'Paso'])['LSPMW'].mean().unstack())
    
    # Mejores y peores configuraciones
    print("\nüèÜ TOP 10 MEJORES CONFIGURACIONES (menor ECRPS promedio):")
    mejores = resultados.groupby(['Tipo_Proceso', 'Proceso', 'Distribuci√≥n', 'Varianza'])['LSPMW'].mean().nsmallest(10)
    print(mejores)
    
    print("\n‚ö†Ô∏è TOP 10 PEORES CONFIGURACIONES (mayor ECRPS promedio):")
    peores = resultados.groupby(['Tipo_Proceso', 'Proceso', 'Distribuci√≥n', 'Varianza'])['LSPMW'].mean().nlargest(10)
    print(peores)
```
