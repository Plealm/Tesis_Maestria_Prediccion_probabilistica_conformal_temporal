```python
import os
import warnings
from pipeline import PipelineExchange
from plot import visualize_predictions

# Filtrar advertencias para limpiar la salida
warnings.filterwarnings("ignore")

def main():
    # =========================================================================
    # CONFIGURACI√ìN
    # =========================================================================
    # El dataset Exchange Rate tiene 8 series correspondientes a diferentes pa√≠ses:
    # 0: Australia, 1: British, 2: Canada, 3: China, 
    # 4: Japan, 5: New Zealand, 6: Singapore, 7: Switzerland
    INDICES_SERIES = [0]  # Puedes cambiar esto a [0, 1, 2...] para correr varias
    
    N_BOOT = 1000         # N√∫mero de muestras para bootstrapping
    SEED = 42             # Semilla para reproducibilidad
    OUTPUT_FOLDER = "Resultados_Exchange_Rate"
    
    print(f"üöÄ Iniciando ejecuci√≥n para Dataset: Exchange Rate")
    print(f"üìÇ Carpeta de salida: {OUTPUT_FOLDER}")
    print(f"‚öôÔ∏è  Configuraci√≥n: √öltimos 1825 d√≠as | Test: √öltimos 30 d√≠as")
    print("-" * 60)

    # =========================================================================
    # INICIALIZAR PIPELINE
    # =========================================================================
    # Instanciamos la clase espec√≠fica para Exchange Rate
    pipeline = PipelineExchange(
        n_boot=N_BOOT, 
        seed=SEED, 
        verbose=True,      # Ver logs de optimizaci√≥n
        val_ratio=0.15     # 15% validaci√≥n
    )

    # =========================================================================
    # EJECUCI√ìN POR SERIE
    # =========================================================================
    for idx in INDICES_SERIES:
        nombre_serie = f"Exchange_Series_{idx}"
        print(f"\nüìà Procesando Serie {idx}...")
        
        try:
            # Llamamos al orquestador visual de plot.py
            # Esto generar√°:
            # 1. Excel con resultados
            # 2. Gr√°ficas Tipo A (Vertical Stack 24+ pasos)
            # 3. Gr√°ficas Tipo B (Corte transversal)
            # 4. Carpeta 'Analisis' con tests estad√≠sticos (Diebold-Mariano)
            visualize_predictions(
                pipeline=pipeline, 
                series_index=idx, 
                output_prefix=nombre_serie, 
                output_folder=OUTPUT_FOLDER
            )
            
            print(f"‚úÖ Serie {idx} completada exitosamente.")
            
        except Exception as e:
            print(f"‚ùå Error procesando la serie {idx}: {str(e)}")
            import traceback
            traceback.print_exc()

    print("\n" + "="*60)
    print("üèÅ EJECUCI√ìN FINALIZADA")
    print("="*60)

if __name__ == "__main__":
    main()
```
